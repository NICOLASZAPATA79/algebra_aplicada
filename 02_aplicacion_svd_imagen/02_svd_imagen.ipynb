# Apliacaci√≥n de SVD  a una imagen

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
image = Image.open('../imagenes/imagen_ejemplo_frida_bredesen.jpg')
plt.imshow(image)
plt.show()
```

```python
imagen_gr = image.convert('LA')
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
imagen_mat.shape = (imagen_gr.size[1],imagen_gr.size[0])
print(imagen_mat.shape)
```

```python
U,D,V = np.linalg.svd(imagen_mat)
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

Las imagenes se pueden reconstruir a partir de la cantidad de informaci√≥n (filas,columnas) que se les brinde.

```python
imagen_rebuild = np.matrix(U[:,:1])*np.diag(D[:1])*np.matrix(V[:1,:])
plt.imshow(imagen_rebuild,cmap='gray')
plt.show()
```

Se puede comprobar con cuantos valores singulares la imagen se empieza a ver como la original, para ello se pueden verificar varios valores

```python
i = 5
imagen_rebuild2 = np.matrix(U[:,:i])*np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_rebuild2,cmap='gray')
title = f'Valores singulares = {i}'
plt.title(title)
plt.show()
```

```python
i = 10
imagen_rebuild3 = np.matrix(U[:,:i])*np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_rebuild3,cmap='gray')
title = f'Valores singulares = {i}'
plt.title(title)
plt.show()
```

```python
i = 50
imagen_rebuild5 = np.matrix(U[:,:i])*np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_rebuild5,cmap='gray')
title = f'Valores singulares = {i}'
plt.title(title)
plt.show()
```

```python
i = 100
imagen_rebuild6 = np.matrix(U[:,:i])*np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_rebuild6,cmap='gray')
title = f'Valores singulares = {i}'
plt.title(title)
plt.show()
```

Se puede observar que al aumentar la cantidad de valores singulares, la imagen empieza a parecerse a la original. Con 50 valores singulares es suficiente para reconocer la imagen. Esto es debido a que son los primeros valores los que poseen mayor varianza, son m√°s representativos.
