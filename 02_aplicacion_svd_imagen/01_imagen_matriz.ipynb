# Una imagen es una matriz

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
```

```python
imagen = Image.open('../imagenes/imagen_ejemplo_frida_bredesen.jpg')
plt.imshow(imagen)
plt.show()
```

```python
imagen_gr = imagen.convert('LA')
print(imagen_gr)
plt.imshow(imagen_gr)
```

Para poder operar con la imagen como un objeto matematico, es necesario obtener los valores nmeros de los pixeles, para asi formar una lista y luego una matriz. 

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
print(imagen_mat)
```

Recordar que los valores de los colores estn entre 0-255.

Para volver el vector en una matriz, utilizamos la funcin **shape**, para darle unas dimensiones

```python
print(imagen_gr.size[1])
print(imagen_gr.size[0])
```

```python
imagen_mat.shape = (imagen_gr.size[1],imagen_gr.size[0])
print(imagen_mat)
print(imagen_mat.shape) # (Filas, columnas)
```

```python
plt.imshow(imagen_mat)
plt.show()
```

```python
plt.imshow(imagen_mat, cmap='gray')
plt.show()
```

```python
imagen_mat2 = imagen_mat/2
print(imagen_mat2)
```

```python
plt.imshow(imagen_mat2,cmap='gray')
plt.show()
```

```python
imagen_mat3 = imagen_mat/100
print(imagen_mat3)
plt.imshow(imagen_mat3,cmap='gray')
plt.show()
```

Se puede observar que al realizar operaciones con los valores que representan los pixeles, cambian las escalas pero no la relacin entre estos, por ello la imagen parece ser la misma .
