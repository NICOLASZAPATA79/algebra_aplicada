# Usando la pseudoinversa 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,1000)
y_1 = -4*x+3
y_2 = 2*x+5
y_3 = -3*x+1

plt.plot(x,y_1,label='y_1')
plt.plot(x,y_2,label='y_2')
plt.plot(x,y_3,label='y_3')

plt.axvline(x=0,color='gray')
plt.axhline(y=0,color='gray')

plt.xlim(-2,2.5)
plt.ylim(-6,6)
plt.legend()
plt.show()
```

Se puede observar que el sistema de ecuaciones es un sistema sobredeterminado ya que posee más variables que ecuaciones, por ello no se pude utilizar el metodo de matrices cuadradas para resolver las ecuaciones asociadas. En este caso se puede utilizar el metodo SVD para resolver el sistema, especificamente utilizar la pseudoinversa de Moore Penrose.

Se quiere hallar la solucion del sistema de ecuaciones Ax = b, tal que x hace que ||Ax-b||_2 sea minima.

```python
matrix = np.array([
    [4,1],
    [-2,1],
    [3,1]
])
print(matrix)
```

```python
matrix_pse= np.linalg.pinv(matrix)
print(matrix_pse)
```

```python
b = np.array([
    [3],
    [5],
    [1]
])
print(b)
```

Para obtener los valores de x,y que relativamente satisfacen el sistema de ecuaiones se realiza la siguiente operación.

```python
result = matrix_pse.dot(b)
print(result)
```

```python
x = np.linspace(-5,5,1000)
y_1 = -4*x+3
y_2 = 2*x+5
y_3 = -3*x+1

plt.plot(x,y_1,label='y_1')
plt.plot(x,y_2,label='y_2')
plt.plot(x,y_3,label='y_3')
plt.scatter(result[0],result[1],label='result')

plt.axvline(x=0,color='gray')
plt.axhline(y=0,color='gray')

plt.xlim(-2,2.5)
plt.ylim(-6,6)
plt.legend()
plt.show()
```

Se puede observar que el punto no queda en la intersección entre las tres rectas, esto es porque al momento de realizar el procedimiento, hay ecuaciones que tienen mayor peso que las otras, lo que hace que los valores que satisfacen el sistema de ecuaciones estén más cercanos a la recta creada con la ecuación de mayor peso.
